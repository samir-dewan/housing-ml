```python
import pandas as pd
import numpy as np
import requests
from zipfile import ZipFile
import io

url= 'https://api.worldbank.org/v2/en/indicator/NY.GDP.MKTP.CD?downloadformat=csv'

response = requests.get(url, stream=True, verify=False)
with ZipFile(io.BytesIO(response.content)) as myzip:
    with myzip.open(myzip.namelist()[1]) as myfile:
        next(myfile)
        data = pd.read_csv(myfile, encoding='utf-8', header=1)

df = pd.DataFrame(data)

df
```

```python
file2 = pd.read_csv('./gdp-per-capita-worldbank.csv', encoding='utf-8', header=0)
df2 = pd.DataFrame(file2)
df2.info()
```

```python
df2[['Code', 'Year', 'GDP per capita, PPP (constant 2017 international $)']]
df2['Year'] = pd.to_datetime(df2['Year'], format='%Y')
recent_indices = df2.groupby('Code')['Year'].idxmax()

most_recent_df = df2.loc[recent_indices]

print(most_recent_df.dtypes)
print(most_recent_df.head(50))
```

```python
print(most_recent_df.dtypes)
print(df.dtypes)
```

```python
df3 = pd.merge(df, most_recent_df, how='left', left_on='Country Code', right_on='Code')
df3['2021'] = pd.to_numeric(df3['2021'], errors='coerce')
df3['GDP per capita, PPP (constant 2017 international $)'] = pd.to_numeric(df3['GDP per capita, PPP (constant 2017 international $)'], errors='coerce')
print(df3.dtypes)
```

```python
df3.head(20).plot.bar(x='Country Name', y=['2021', 'GDP per capita, PPP (constant 2017 international $)'], secondary_y='2021')
```

```python
file3 = pd.read_csv('./countries-table.csv', encoding='utf-8', header=0)
countries_df = pd.DataFrame(file3)
countries_df.info()
countries_df[['country', 'pop2023']].head(10)
```

```python
df4 = pd.merge(df3, countries_df, how='left', left_on='Country Name', right_on='country')
df4[['Country Name', 'pop2023', '2021', 'GDP per capita, PPP (constant 2017 international $)']]
df4['co2 per capita'] = df4['2021']/df4['pop2023']
df4[['Country Name', 'co2 per capita', 'GDP per capita, PPP (constant 2017 international $)']].dtypes
df4.head(20).plot.bar(x='Country Name', y=['GDP per capita, PPP (constant 2017 international $)', 'co2 per capita'], secondary_y='co2 per capita')
```

```python
df4_nona = df4[df4["co2 per capita"].notna()]
df4_nona = df4_nona[df4_nona['GDP per capita, PPP (constant 2017 international $)'].notna()]
df4_nona.info()
df4_nona.head(20).plot.bar(x='Country Name', y=['GDP per capita, PPP (constant 2017 international $)', 'co2 per capita'], secondary_y='co2 per capita')
```

```python
# df4_nona.plot.scatter(x="co2 per capita", y="GDP per capita, PPP (constant 2017 international $)")
df4_nona['log_co2pc'] = np.log(df4_nona['co2 per capita'])
df4_nona['log_gdppc'] = np.log(df4_nona['GDP per capita, PPP (constant 2017 international $)'])
df4_nona.plot.scatter(x="log_co2pc", y="log_gdppc")
```
